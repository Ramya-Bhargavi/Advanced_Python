<h1 style="color:blue">Pivot basics</h1>

Pivot and PivotTable are an interactive way to quickly summarize large amounts of data.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/sample_data/weather.csv")
df
```

here the data set is not having same values for combination of different columns.

```python
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='city',columns='date',values="humidity")
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='humidity',columns='city')
```

<h1 style="color:blue">Pivot Table</h1>

data set when it has two similar rows values.

```python
df = pd.read_csv("/content/sample_data/weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date",aggfunc="sum")
```

```python
df.pivot_table(index="city",columns="date", aggfunc=np.sum)  
```

The basic difference between pivot() and pivot_table() are given below: 

pivot_table() does not raise error for multiple entries of a row, 
column combination. 

It aggregate the multiple entries present for a row-column combination.

```python
df.pivot_table(index="city",columns="date")  #by default the aggfunc it is mean  (80+83)/2 = 81.5
```

<h2 style="color:brown">Margins</h2>

```python
df.pivot_table(index="city",columns="date", margins=True)   #the sum of city values is showing on all column and teh dates values sum on all column
```

<h2 style="color:brown">Grouper</h2>

A Grouper allows the user to specify a groupby instruction for an object.

```python
df = pd.read_csv("/content/sample_data/weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
type(df['date'][0])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')  # freq= M = month --- monthly avg temperature
# on date we are applying grouping -- key
#we are displaying the columns as city
```
