<h1 align="center" style="color:green">Matplotlib Tutorial: Axes Labels, Legend, Grid</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]  # weeks 1 -7
max_t=[50,51,52,48,47,49,46] # maximum tenperature list recorded in the corresponding weeks
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# in the same plot we are adding three different lines
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")


# after the chart we do not know what color is representing which line and which plot
```

<h2 style="color:purple">Axes labels and chart title</h2>

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

#still we do not know what is which line
```

<h2 style="color:purple">Legend</h2>

```python
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best') # just the legend will add the legends like what colors are what labels and the 
 #loc=best tells where the legend has to be placed in matplotlib so it looks good it will decide for itself
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

<h2 style="color:purple">Grid</h2>

```python
#to exactly see it as a graph we can use grid lines
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right', fontsize="large",shadow=True)

plt.grid()
```
