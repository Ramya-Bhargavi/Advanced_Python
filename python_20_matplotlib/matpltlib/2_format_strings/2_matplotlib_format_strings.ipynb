<h1 align="center" style="color:green">Matplotlib Tutorial: pyplot.plot format strings and other options</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--') # b is for blue and drawing in + -- format.... here b+ is like format string
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(x,y,"+-r")
```

```python
#error for not giving proper format string
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(x,y,"+-D")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(x,y,"rd--")  # d -- diamond
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

<h3 style="color:purple">markersize</h3>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

<h3 style="color:purple">alpha property to control transparency of line chart</h3>

```python
plt.plot(x,y,color='b',alpha=1) # alpha can be specified on a scale 0 to 1
```

```python
plt.plot(x,y,alpha=0.5)
```

```python
plt.plot(x,y,'g<',alpha=1)
```
