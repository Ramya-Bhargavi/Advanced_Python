<h1 align="center", style="color:green">Matplotlib Tutorial: Bar Chart</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Simple bar chart showing revenues of major tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
x = np.arange(len(company))  # to get the numbers same as list number of elements
x
```

```python
plt.bar(x,revenue, label="Revenue")
```

nothing in the above chart is clear what are the bars representing so we use xticks, we can have x label but it is not required as we have to explain what each bar is...

```python
plt.bar(x,revenue, label="Revenue")

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

<h3>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
plt.bar(x,revenue, label="Revenue") 
plt.bar(x,profit,label="Profit")

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(x-0.1,revenue, label="Revenue") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 
plt.bar(x+0.1,profit,label="Profit") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.grid()
```

```python
plt.bar(x-0.2,revenue, label="Revenue") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 
plt.bar(x+0.2,profit,label="Profit") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(x-0.2,revenue,width=0.2,label="Revenue") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 
plt.bar(x+0.2,profit,width=0.2,label="Profit") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(x-0.2,revenue,width=0.4,label="Revenue") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 
plt.bar(x+0.2,profit,width=0.4,label="Profit") # if x is at 0 then from -0.2 we start the revenue bar and  from 0.2 

plt.xticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

<h3>Horizontal bar chart using barh function<h3>

```python
plt.barh(x,revenue,label="Revenue")

plt.yticks(x,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
