### Correlation with Heatmap

 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

Correlation is used to find the relationship between different features only for the numeric values like integer and float.

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. 

It can be usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot. Do ctrl+tab to check params.


In above heat map we have seen that the corr between the bill and tip is more high, so we are finding out individual relationship between two features only.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')  # kind paramter is used to mention the shape of the points in the map

# where the histogram heights are more i.e. frequency if tips is more there, then it means that more tips are given in that particular region
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')  # reg--- regression 
#it also draws a probability density function
```

## Pair plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df) # here for all different features the permutations and combinations will be carried out and the for the analysis of same features like tip to tip or bill to bill the frequency of that feature will be shown 

# for other bivariate analysis the scatter plots will be shown
```

All the plots inside the pair plots can also be classified based on the feature using hue parameter.

```python
sns.pairplot(df,hue='sex') # by observing in row of tip and total bill there are more blue dots means more tips are by males
```

```python
sns.pairplot(df,hue='smoker') # by observing in row of tip and total bill there are more orange dots means more tips are by non smokers
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde="false") # here we have density by defult kde=true that why on y axis we are seeing density values. 

#Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
## Count plot

sns.countplot(x='sex',data=df)
```

```python
## Count plot

sns.countplot(y='sex',data=df)
```

```python
## Bar plot
sns.barplot(x='total_bill',y='sex',data=df) # for two different features
```

```python
## Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker',y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
